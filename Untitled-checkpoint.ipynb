```python
fiction_books = bestsellers_df[bestsellers_df['Genre'] == 'Fiction']
fiction_average_rating_by_year = fiction_books.groupby('Year')['User Rating'].mean()
fiction_average_rating_by_year.plot(kind='line', marker='o')
plt.xlabel('Year')
plt.ylabel('Average User Rating')
plt.title('Average User Rating for Fiction Books Over Time')
plt.show()
```

```python
# Scatter plot to visualize the relationship between price and user rating
plt.scatter(bestsellers_df['Price'], bestsellers_df['User Rating'], alpha=0.5)
plt.xlabel('Price')
plt.ylabel('User Rating')
plt.title('Relationship between Price and User Rating')
plt.show()

# Scatter plot to visualize the relationship between price and reviews
plt.scatter(bestsellers_df['Price'], bestsellers_df['Reviews'], alpha=0.5)
plt.xlabel('Price')
plt.ylabel('Number of Reviews')
plt.title('Relationship between Price and Number of Reviews')
plt.show()

# Correlation matrix to identify numerical relationships
correlation_matrix = bestsellers_df[['Price', 'User Rating', 'Reviews']].corr()
```

```python
# Calculate the average user rating for each author across all years
avg_rating_by_author = bestsellers_df.groupby('Author')['User Rating'].mean()
top_authors = avg_rating_by_author.nlargest(3)

# Extract summary information for the top authors
top_author_summary = bestsellers_df[bestsellers_df['Author'].isin(top_authors.index)]
```

```python
top_author_summary
```

```python
from scipy.stats import ttest_ind

fiction_prices = bestsellers_df[bestsellers_df['Genre'] == 'Fiction']['Price']
non_fiction_prices = bestsellers_df[bestsellers_df['Genre'] == 'Non Fiction']['Price']

t_statistic, p_value = ttest_ind(fiction_prices, non_fiction_prices)
```
